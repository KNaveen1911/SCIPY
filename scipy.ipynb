```python
s=sp.rand(10)
```

```python
print(s)
```

```python
print(len(s))
```

```python
print("mean:",s.mean())
```

```python
print("mean:{0:8.3f}".format(s.mean()))
```

```python
print(s.min())
```

```python
print("minimum:{0:8.2f}".format(s.min()))
```

```python
print(s.max())
```

```python
print("maximum:{0:8.2f}".format(s.max()))
```

```python
#it measures how far a set of (random)numbers are spread out from their average value
print("variance:{0:8.2f}".format(s.var()))
```

```python
# square root of the variance
print("std.deviation:{0:8.2f}".format(s.std()))
```

```python
import scipy as sp
from scipy import integrate
from scipy import special
```

```python
# special function--exponential

x=special.exp10(4)
print(x)
```

```python
x=special.exp10([2,3,4])
print(x)
```

```python
x=special.exp2(3)
print(x)
```

```python
x=special.exp2([2,4,6,8])
print(x)
```

### special functions-- trigonpmetric functions

```python
a=special.sindg(90)
print(a)
```

```python
a=special.cosdg(30)
a
```

```python
a=special.tandg(60)
a
```

```python
a=special.cotdg(90)
a
```


# integral- general ( quad)

```python
i=sp.integrate.quad(lambda x:special.exp10(x),0,1)
i   # gives 2 values real and imaginary
```

```python
a=lambda x:((x**2)-4*x + 5)
i=integrate.quad(a,0,1)
i
```

### double integration--dblquad

```python
a=lambda x,y:x*y**3  
#a=lambda x,y:(x*y)**3   
b=1
c=-1
i=integrate.dblquad(a,0,2,b,c)
i
```
